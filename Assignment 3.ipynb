### Q4a

```r
fc18scores <- c(0:8)
fc18freq <- c(1,15,17,19,5,2,2,3,0)
fc18games <- sum(fc18freq)
```

```r
q4df <- data.frame( 
               goals = rep( fc18scores
                          , times = fc18freq))
plot(fc18scores, fc18freq, type='h')
hist(q4df$goals)
```

```r
theta4a = sum(q4df$goals)/fc18games
theta4a
```

```r
theta4a/fc18games
```

```r
plot(fc18scores, fc18games * dpois(fc18scores, lambda = theta4a))
```
